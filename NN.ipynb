```python
#NN-2-2-2-1-momentum-batch
import numpy as np 
#np.random.seed(0)

def sigmoid (x):
    return 1/(1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

#Input datasets
inputs = np.array([[0,0],[0,1],[1,0],[1,1]])
expected_output = np.array([[0],[1],[1],[0]])

epochs = 30000
lr = 0.2
alpha = 0.3
In, H1, H2, out = 2,4,4,1

#Random weights and bias initialization
H1_weights = np.random.uniform(size=(In,H1))
H1_bias =np.random.uniform(size=(1,H1))
H2_weights = np.random.uniform(size=(H1,H2))
H2_bias =np.random.uniform(size=(1,H2))
out_weights = np.random.uniform(size=(H2,out))
out_bias = np.random.uniform(size=(1,out))

H1_prev_weights = np.zeros([In,H1])
H2_prev_weights = np.zeros([H1,H2])
out_prev_weights = np.zeros([H2,out])


print("Initial hidden-1 weights: ",end='')
print(*H1_weights)
print("Initial hidden-1 biases: ",end='')
print(*H1_bias)
print("Initial hidden-2 weights: ",end='')
print(*H2_weights)
print("Initial hidden-2 biases: ",end='')
print(*H2_bias)
print("Initial output weights: ",end='')
print(*out_weights)
print("Initial output biases: ",end='')
print(*out_bias)


#Training algorithm
for iter in range(epochs):
  #Forward Propagation
  H1_activation = np.dot(inputs,H1_weights)
  H1_activation += H1_bias
  H1_output = sigmoid(H1_activation)

  H2_activation = np.dot(H1_output,H2_weights)
  H2_activation += H2_bias
  H2_output = sigmoid(H2_activation)

  out_activation = np.dot(H2_output,out_weights)
  out_activation += out_bias
  predicted_output = sigmoid(out_activation)
  #print(*predicted_output)
  
  #Backpropagation
  error = expected_output - predicted_output
  if(iter%1000==0):
    print(f"loss: {np.mean(np.square(error))}")
  #print(*error)
  d_predicted_output = error * sigmoid_derivative(predicted_output)

  error_H2 = d_predicted_output.dot(out_weights.T)
  d_H2 = error_H2 * sigmoid_derivative(H2_output)

  error_H1 = d_H2.dot(H2_weights.T)
  d_H1 = error_H1 * sigmoid_derivative(H1_output)


  #Updating Weights and Biases
  out_weights += H2_output.T.dot(d_predicted_output) * lr + out_prev_weights * alpha
  out_prev_weights = H2_output.T.dot(d_predicted_output) * lr + out_prev_weights * alpha
  out_bias += np.sum(d_predicted_output,axis=0,keepdims=True) * lr

  H2_weights+= H1_output.T.dot(d_H2)*lr + H2_prev_weights *alpha
  H2_prev_wights = H1_output.T.dot(d_H2)*lr + H2_prev_weights *alpha
  H2_bias += np.sum(d_H2,axis=0,keepdims=True)*lr

  H1_weights += inputs.T.dot(d_H1) * lr + H1_prev_weights * alpha
  H1_prev_weights = inputs.T.dot(d_H1) * lr + H1_prev_weights * alpha
  H1_bias += np.sum(d_H1,axis=0,keepdims=True) * lr

print("Initial hidden-1 weights: ",end='')
print(*H1_weights)
print("Initial hidden-1 biases: ",end='')
print(*H1_bias)
print("Initial hidden-2 weights: ",end='')
print(*H2_weights)
print("Initial hidden-2 biases: ",end='')
print(*H2_bias)
print("Initial output weights: ",end='')
print(*out_weights)
print("Initial output biases: ",end='')
print(*out_bias)

print("\nOutput from neural network after 10,000 epochs: ",end='')
print(*predicted_output)
```
